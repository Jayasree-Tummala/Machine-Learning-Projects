# **We are performing univariate Analysis**

```python
import pandas as pd
import seaborn as sns

```

 # Loading Titanic Dataset**

df = pd.read_csv("titanic.csv")

```python
df.sample(5)
```

# **Visualising Categorical data like Gender, Sex, Pclass**
## *For categorical data we can use bar charts, Pie charts to represent them*

```python
print(df['Survived'].unique())
```

```python
sns.countplot(x='Survived', data=df)
```

```python
df['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
```

# **Numerical data Visualization**

## *Histogram*

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=10)
```

## *Distplot*

```python
sns.distplot(df['Age'])
```

## *Boxplot*

```python
sns.boxplot(df['Fare'])
```
